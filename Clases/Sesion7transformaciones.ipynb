```python
# sample: es una transformacion que devuelve un conjunto de filas seleccionadas
# aleatoriamente del dataframe el numero de filas de vuelta sera aproximadamente
# al numero especificado que representa a un porcentaje y a un valor que
# representa entre 0 y 1
df_muestra = df.sample(0.8)
```

```python
num_filas = df.count()
num_filas_muestra = df_muestra.count()

print('El 80% de filas del dataframe origianal es {}'.format(num_filas -( num_filas*0.2)))
print('El numero de filas del dataframe muestra es {}'.format(num_filas_muestra))
```

```python
df_muestra = df.sample(fraction =0.8, seed=1234)
```

```python
df_muestra = df.sample(withReplacement=True, fraction =0.8, seed=1234)
```
