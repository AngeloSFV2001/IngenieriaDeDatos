```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
rdd = sc.parallelize([2,4,6,8])
```

```python
rdd.reduce(lambda x,y: x+y)
```

```python
rdd1 = sc.parallelize([1,2,3,4])
```

```python
rdd1.reduce(lambda x,y: x*y)
```

```python
rdd = sc.parallelize(['j','o','s','e'])
```

```python
rdd.count()
```

```python
rdd1 = sc.parallelize(item for item in range(10))
```

```python
rdd1.count()
```

```python
rdd2 = sc.parallelize(range(10))
```

```python
rdd2.count()
```

```python
rdd1.collect()
```

```python
rdd2.collect()
```

```python
rdd3 = sc.parallelize('Hola Apache Spark:'.split(' '))
```

```python
rdd3.collect()
```

```python
rdd4 = sc.parallelize([(item,item ** 2)for item in range(20)])
```
