# Map

```python
rdd = sc.parallelize([1,2,3,4,5])
```

```python
rdd_resta = rdd.map(lambda x: x-1)
rdd_resta.collect()
```

```python
rdd_par = rdd.map(lambda x: x%2==0)
rdd_par.collect()
```

```python
rdd_texto = sc.parallelize(["jose", "juan", "lucia", "angelo"])
rdd_mayus = rdd_texto.map(lambda x: x.upper())
rdd_mayus.collect()
```

```python
rdd_hola = rdd_texto.map(lambda x: "hola "+x)
rdd_hola.collect()
```

# Flapmap

```python
rdd = sc.parallelize([1,2,3,4,5])
```

```python
rdd_cuadrado = rdd.map(lambda x: (x,x**2))
rdd_cuadrado.collect()
```

```python
rdd_cuadrado_flat = rdd.flatMap(lambda x: (x,x**2))
rdd_cuadrado_flat.collect()
```

```python
rdd_texto = sc.parallelize(['jose','juan','lucia','angelo'])
rdd_mayuscula = rdd_texto.map(lambda x: (x, x.upper()))
rdd_mayuscula.collect()
```

```python
rdd_texto = sc.parallelize(['jose','juan','lucia','angelo'])
rdd_mayuscula_flat = rdd_texto.flatMap(lambda x: (x, x.upper()))
rdd_mayuscula_flat.collect()
```

# Filter

```python
rdd= sc.parallelize([1,2,3,4,5,6,7,8,9])
```

```python
rdd_par=rdd.filter(lambda x: x%2==0)
rdd_par.collect()
```

```python
rdd_impar=rdd.filter(lambda x: x%2!=0)
rdd_impar.collect()
```

```python
rdd_texto = sc.parallelize(['jose','juan','lucia','angelo','karla','katia'])
rdd_k = rdd_texto.filter(lambda x: x.startswith('k'))
rdd_k.collect()
```

```python
rdd_filtro =  rdd_texto.filter(lambda x: x.startswith('j') and x.find('u')==1)
rdd_filtro.collect()
```

# reduceByKey


```python
rdd= sc.parallelize(
    [('casa',2),
     ('parque',1),
     ('que',5),
     ('casa',1),
     ('escuela',2),
     ('casa',1),
     ('que',1)])
```

```python
rdd.collect()
```

```python
rdd_reduce = rdd.reduceByKey(lambda x,y: x+y)
```

```python
rdd_reduce.collect()
```
