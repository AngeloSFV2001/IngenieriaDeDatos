```python
factorial_result = sc.parallelize(range(1, 6))
factorial_result.reduce(lambda x, y: x * y)
```

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    rdd = sc.parallelize(list(range(1, n + 1)))
    return rdd.reduce(lambda x, y: x * y)
```
